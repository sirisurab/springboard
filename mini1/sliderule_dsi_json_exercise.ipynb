## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

## load json into dataframe and look at first 5 rows

```python
df = pd.read_json('data/world_bank_projects.json')
df.head()
```

# Find the 10 countries with most projects

```python
df_10mp = df.countryname.value_counts().head(10)
df_10mp
```

```python
#plot histogram of top 10 countries
df_10mp.sort_index().plot(kind='bar')
```

# Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
#investigate the contents of mjtheme_namecode column
proj_themes = df[['mjtheme_namecode','id']]
proj_themes.head()
```

```python
#reload json to invoke normalize on it using mtheme_namecode and id columns (there can be multiple themes for each project)
with open('data/world_bank_projects.json') as file:
    json_f = json.load(file)
l_df = json_normalize(json_f,'mjtheme_namecode',['id'])
l_df.head()
```

```python
#code is string, convert to numeric and filter rows with code values 1-10
l_df['code'] = pd.to_numeric(l_df['code'],errors='coerce')
l_df = l_df[l_df['code'] < 11]
l_df.head()
```

```python
#create code to name map table (as legend for plot below as well as for join in next step)
x_labels= l_df[l_df['name'] != ''][['code','name']].sort_values('code').dropna().drop_duplicates()#['name']
x_labels
```

```python
l_df['code'].value_counts().sort_index().plot(kind='bar')
```

# In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#join l_df with the x_labels dataframe above to create a new column 'name_y' that has the correct names and no null values
new_ldf = pd.merge(l_df,x_labels,on='code',how='inner',suffixes=('', '_y'))
new_ldf.head()
```

```python
#overwrite name_y column to name and drop name_y
new_ldf['name'] = new_ldf['name_y']
new_ldf = new_ldf.drop(columns=['name_y'])
new_ldf.head(15)
```

```python
#bar plot for number of projects for each name. to show number of projects with no name
l_df.name.value_counts().sort_index().plot(kind='bar')
```

```python
#bar plot for number of projects for each name. new_ldf should have 0 projects with no name
new_ldf.name.value_counts().sort_index().plot(kind='bar')
```
